## 3.GPA Calculator

```python
def read_csv(path):
  '''
    data return {subject:[]  , credits:[]   , grade: []}
  '''
  with open(path) as f:
    #clean lines
    lines = [i.strip() for i in f.read().splitlines() if i.strip()!='']
    temp_dict = {}
    header = ['subject','credits','grade']
    #create dict to get data
    for i in range(len(header)):
      temp_dict[header[i]] = list()
    #get data from lines
    for line in lines:
      data = line.split(',')
      for i in range(len(header)):
        temp_dict[header[i]].append(data[i])
    return temp_dict


def grade_to_point(grade):
  if grade == 'F':
    return f' point: 0.0',0
  elif grade == 'D':
    return f' point: 1.0',1
  elif grade == 'D+':
    return f'point: 1.5',1.5
  elif grade == 'C':
    return f' point: 2.0',2
  elif grade == 'C+':
    return f'point: 2.5',2.5
  elif grade == 'B':
    return f' point: 3.0',3
  elif grade == 'B+':
    return f'point: 3.5',3.5
  elif grade == 'A':
    return f' point: 4.0',4



def find_gpa(credits,points_list):
  temp = 0
  for i in range(len(credits)):
    temp+= credits[i]*points_list[i]
  return temp/sum(credits)

def main():
  path = input('Enter grade file: ')
  points_list = []
  header = ['subject','credits','grade']
  data = read_csv(path)
  for i in range(len(data.get(header[0]))):
    s=''
    for head in header:
      s+= head + ': ' + data[head][i] + ' '
    s_tmp,point = grade_to_point(data['grade'][i])
    points_list.append(point)
    s+= s_tmp
    print(s)

  credits = [int(i) for i in data['credits']]
  total_credits = sum(credits)
  print(f'Total credits = {total_credits}')
  print(f'GPA = {find_gpa(credits,points_list):.2f}')
main()
```

## 9.L12 Matrix

```python
class Matrix():
    all_matrix = []
    def __init__(self,matrixx=None):
      self.matrix = matrixx
      Matrix.all_matrix.append(self)

    @staticmethod
    def zero_by_matrix(A):
      zero_ma = []
      for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
          row.append(0)
        zero_ma.append(row)
      return zero_ma

    @staticmethod
    def zero_by_row(row,y):
      zero_ma = []
      for i in range(row):
        row_ = []
        for j in range(y):
          row_.append(0)
        zero_ma.append(row_)
      return zero_ma

    @staticmethod
    def plus_matrix(A,B):
      C = Matrix.zero_by_matrix(A)
      for i in range(len(A)):
        for j in range(len(A[i])):
          C[i][j] = A[i][j] + B[i][j]
      return C

    @staticmethod
    def mul_matrix(A,B):
        mat=Matrix.zero_by_row(len(A),len(B[0]))
        for i in range(0,len(A)):
            for j in range(0,len(B[0])):
                for k in range(0,len(B)):
                    mat[i][j]+=A[i][k]*B[k][j]
        return mat

    
def print_matrix(res):
  for i in range(len(res)):
    for j in range(len(res[0])):
        print(f"{res[i][j]:^5}",end=" ")
    print()

def read_matrix(path):
  with open(path) as f:
    temp = []
    
    data = f.read().splitlines()
    i=0
    while i<len(data):
      row = data[i].split(' ')
      try:
        row = [int(i) for i in row]
      except:
        pass
      if len(row)>1:
        temp.append(row)
        i+=1
      else:
        Matrix(temp)
        temp = []
        i+=1
    Matrix(temp)


    
path = input('File name: ')
read_matrix(path)
all_matrix = Matrix.all_matrix
mat1 = Matrix.mul_matrix(all_matrix[0].matrix,all_matrix[1].matrix)
mat2 = Matrix.plus_matrix(mat1,all_matrix[2].matrix)
print_matrix(mat2)
```

## 10.L12 Buy list

```python
def get_price(path):
    temp = {}
    with open(path) as f:
        data = f.read().splitlines()
        is_price = False
        for line in data:
            if line == 'Price':
                is_price = True
            if is_price:
                if line == 'List':
                    is_price = False
                elif len(line.split(' ')) == 2:
                    item,price = line.split(' ')
                    temp[item] = int(price)               
    return temp

def get_order(item_list,path):
    price_total = 0
    with open(path) as f:
        data = f.read().splitlines()
        is_list = False
        for line in data:
            if line == 'List':
                is_list = True
            if is_list:
                if line == 'Price':
                    is_list = False
                elif len(line.split(' ')) == 2:
                    item,amount = line.split(' ')
                    if item_list.get(item) != None:
                        price_total += item_list[item]*int(amount)
    return price_total

def print_data(path):
    with open(path) as f:
        data = f.read().splitlines()
        for line in data:
            print(line)

path = input('File name: ')
print_data(path)
item_list = get_price(path)
price = get_order(item_list,path)
print(f'Total price: {price}')
```

## 14.L12 GcdLcm

```python
def is_prime(n):
    for i in range(2,(n//2)+1):
        if n%i==0:
            return False
    return True


def find_factor(number):
    s = list()
    if number == 1:
        return [1]
    elif is_prime(number):
        return [1,number]
    else:
        for i in range(2,(number//2)+1):
            if number%i==0 and is_prime(i):
                pow_ = 1
                while number%(i**pow_)==0:
                    pow_+=1
                pow_-=1
                for time in range(pow_):
                    s.append(i)
    return s


def find_gcd(n1,n2):
    factor1 = find_factor(n1)
    factor2 = find_factor(n2)
    gcd = 1
    for f1 in factor1:
        if f1 in factor2:
            ind = factor2.index(f1)
            gcd*=factor2.pop(ind)
    return gcd

def find_lcm(n1,n2):
    gcd = find_gcd(n1,n2)
    multi = n1*n2
    return (n1*n2)/gcd

def main():
    a = int(input('a : '))
    b = int(input('b : '))
    gcd = find_gcd(a,b)
    lcm = find_lcm(a,b)
    print(f'GCD : {gcd}')
    print(f'LCM : {lcm:.0f}')

main()
        
```

## 15.L12 mine_sweeper

```python
def create_grid(x,y):
    grid_list = list()
    for j in range(y):
        k=list()
        for i in range(x):
            k.append(0)
        grid_list.append(k)
    return grid_list

def place_mine(grid_list):
    n=int(input('Number of mine(s): '))
    for time in range(n):
        pos_x,pos_y = input(f'Mine#{time+1}: ').split(' ')
        pos_x,pos_y = int(pos_x),int(pos_y)
        grid_list[pos_y][pos_x] = 'X'
        max_y,low_y =  pos_y+1,pos_y-1
        max_x,low_x = pos_x+1,pos_x-1        
        for j in range(len(grid_list)):
            for i in range(len(grid_list[0])):
                if i in range(low_x,max_x+1) and j in range(low_y,max_y+1):
                    if grid_list[j][i]!='X':
                        grid_list[j][i] +=1
    return grid_list

def print_result(grid_list):
    for row in grid_list:
        row = [str(i) for i in row]
        print(' '.join(row))


def main():
    x,y = input('Grid Size: ').split(' ')
    x,y = int(x),int(y)
    grid_list = create_grid(x,y)
    print_result(place_mine(grid_list))
main()
```

## 16.L12 word_chain

```python
def main():
    word_list = input('Text: ').split(' ')
    s=''
    all_chain = list()
    nchain = 0
    length = len(word_list[0])
    to_pass = length-2
    combo = 0
    for word in word_list:
        if s=='':
            s=word
            all_chain.append([word])
            nchain+=1
        elif s!='':
            combo = 0
            for i in range(length):
                if s[i] == word[i]:
                    combo+=1
            if combo < to_pass:
                all_chain.append([word])
                s=word
                nchain+=1
            else:
                all_chain[nchain-1].append(word)
                s=word

    max_length = max([len(i) for i  in all_chain])
    print(f'{len(all_chain)} Chain(s). Maximum length is {max_length} word(s).')           

main()
```
