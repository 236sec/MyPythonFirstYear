## 1.L14 Student

```python
import csv
maxx = -1
minn = 1e9
def read_csv(filename):
    with open(filename) as f:
        global maxx , minn
        datas = csv.reader(f)
        data = [ i for i in datas]
        for i in range(1,len(data)) :
            maxx = max(maxx,int(data[i][6]))
            minn = min(minn,int(data[i][7]))
    return data
```

```python
class Std():
    student_list = dict()
    def __init__(self,id,gender,race_ethnicity,parental,lunch,test_preparation_course,math_score,reading_score,writing_score):
        self.id = id
        self.gender = gender
        self.race_ethnicity = race_ethnicity
        self.parental = parental
        self.lunch = lunch
        self.test_preparation_course = test_preparation_course
        self.math_score = int(math_score)
        self.reading_score = int(reading_score)
        self.writing_score = int(writing_score)
        pass

    def show_data(self):
        print(f'''gender : {self.gender}
race/ethnicity : {self.race_ethnicity}
parental level of education : {self.parental}
lunch : {self.lunch}
test preparation course : {self.test_preparation_course}
math score : {self.math_score}
reading score : {self.reading_score}
writing score : {self.writing_score}''')

    def less_from_max_math_score(self):
        less = maxx - self.math_score
        print(f'less than max : {less}')
    def more_from_min_reading_score(self):
        more = self.reading_score - minn
        print(f'more than min : {more}')
    def can_pass(self):
        score_total = self.reading_score + self.math_score
        if score_total > 240:
            print('Pass')
        else:
            print('Can\'t pass')
    def __str__(self):
        math_score = self.math_score
        read_score = self.reading_score
        write_score = self.writing_score
        return f'math score : {math_score}, reading score : {read_score}, writing score : {write_score}'
```

```python
filename = input("Filename : ")
datas = read_csv(filename)
classes = list()
for i in range(1,len(datas)) :
    data = Std(datas[i][0],datas[i][1],datas[i][2],datas[i][3],datas[i][4],datas[i][5],datas[i][6],datas[i][7],datas[i][8])
    classes.append(data)
menu = int(input("Menu : "))
id = input("ID : ")
for i in classes :
    if i.id == id :
        if menu == 1 : # show data
            i.show_data()
        elif menu == 2 : # less_from_max_math_score
            i.less_from_max_math_score()
        elif menu == 3 : # more_from_min_reading_score
            i.more_from_min_reading_score()
        elif menu == 4: # can_pass
            i.can_pass()
        elif menu == 5: # print
            print(i)
        break
```

## 2.L14 Account

```python
import json

class classes():
    customer_id = dict()
    def __init__(self,data = dict()):
        self.name = data.get('name',None)
        self.id = data.get('account number',None)
        self.money = data.get('money',None)
        self.history = data.get('history')
        if self.history == '':
            self.history = dict()
        classes.customer_id[self.id] = self
    
    def show_data(self):
        print(f'Name : {self.name}')
        print(f'Account number : {self.id}')
        print(f'Money : {self.money}')
        de_po = self.history.get('Deposit',False)
        wi_aw = self.history.get('Withdraw',False)
        print_depo = False
        if de_po:
            print(f'History : Deposit : {de_po}')
            print_depo = True
        if wi_aw:
            if print_depo:
                print(f'Withdraw : {wi_aw}')
            else:
                print(f'History : Withdraw : {wi_aw}')
        if not(de_po) and not(wi_aw):
            print('History :')

    def deposit(self):
        amount = int(input('Amount : '))
        self.money+=amount
        if 'Deposit' in self.history.keys():
            self.history['Deposit'] += amount
        else:
            self.history['Deposit'] = amount
        print(f'Current money : {self.money}')

    def withdraw(self):
        amount = int(input('amount : '))
        self.money -=amount
        if 'Withdraw' in self.history.keys():
            self.history['Withdraw'] += amount
        else:
            self.history['Withdraw'] = amount
        print(f'Current money : {self.money}')

    def show_history(self):
        if self.history.get('Deposit',False):
            de_po = self.history.get('Deposit')
            print(f'Deposit : {de_po}')
        if self.history.get('Withdraw',False):
            wi_aw = self.history.get('Withdraw')
            print(f'Withdraw : {wi_aw}')

def get_json(path):
    with open(path,'r') as f:
        data = f.read()
        data = json.loads(data)
    return data
        
def main():
    path = input('Filename : ')
    data = get_json(path)
    for cus in data:
        classes(cus)
    while True:
        ask_id = input('Account number : ')
        while True:
            op = input('Menu : ')
            customer = classes.customer_id[ask_id]
            if op == '0':
                break
            elif op == '1':
                customer.show_data()
            elif op=='2':
                customer.deposit()
            elif op=='3':
                customer.withdraw()
            elif op=='4':
                customer.show_history()
            elif op=='5':
                break
        if op=='0':
            break

main()
```

## 3.L14 Racing

```python
import json
def get_data(path):
    with open(path,'r') as f:
        data = json.loads(f.read())
    return data

def velocity_range(velo,_range):
    time = _range/velo
    return time

def find_speed(data,meter):
    temp = dict()
    for person in data:
        _range = [float(i) for i in person.get('_range')]
        velo_list = [float(i) for i in person.get('velocity')]
        if _range == [0]:
            time = velocity_range(velo_list[0],meter)
            time = float(f'{time:.2f}')
            res = int(person['response'])
            name = person['name']
            temp[name] = (time,-res,name)
        else:
            s_total = 0
            time = 0
            i=1
            while i < len(_range):
                if _range[i] <= meter:
                    s = _range[i] - _range[i-1]
                    s_total+=s
                    time+= velocity_range(velo_list[i-1],s)
                elif _range[i] > meter:
                    s = _range[i] - meter
                    time+= velocity_range(velo_list[i-1],s)
                    s_total+=s
                    break
                i+=1
            if s_total != meter:
                s = meter - _range[-1]
                time+=velocity_range(velo_list[-1],s)
            time = float(f'{time:.2f}')
            name = person['name']
            res = int(person['response'])
            temp[name] = (time,-res,name)
    return temp

def print_scoreboard(person_dict):
    time_sort = sorted(list(person_dict.values()),reverse=False)
    for time in time_sort:
        for person in person_dict:
            if person_dict[person] == time:
                print(f'{person} {person_dict[person][0]:.2f}')
                break

def main():
    path = input('Enter json filename : ')
    #path = 'python\data\\racing2.json'
    data = get_data(path).get('racer')
    meter = int(input('Enter distance (meter) : '))
    score = find_speed(data,meter)
    print_scoreboard(score)

main()
```

## 4.L14 Item

```python
class Item():
    item_dict = dict()
    def __init__(self,ID=str(),type=str(),price=int()):
        self.ID = ID
        self.type = type
        self.price = price
        Item.item_dict[ID] = self

def get_input():
    n=int(input('How many products are there? : '))
    for i in range(n):
        item_id,item_type,item_price = input().split(' ')
        Item(item_id,item_type,item_price)
 

def main():
    get_input()
    while True:
        id_ask = input('Id : ')
        if not(id_ask in Item.item_dict):
            print('This id doesn\'t exist.')
            continue
        else:
            item = Item.item_dict.get(id_ask)
            while True:
                op = input('Q : ')
                if op == '0':
                    break
                elif op == '1':
                    print(item.type)
                elif op == '2':
                    print(item.price)
                elif op == '3':
                    break
        if op=='0':
            break

main()
```

## 9.L14 World population

```python
import json
def get_data(path):
    with open(path,'r') as f:
        data = json.loads(f.read()) 
    return data

def get_total_population(data):
    su = 0
    for country in data:
        su += int(country.get('population'))
    return su

def fliter_population(data):
    su_1,su_2,su_3 = 0,0,0
    for country in data:
        popu = int(country.get('population'))
        if popu > 500*(10**6):
            su_1+=1
        if 250*(10**6)< popu < 750*(10**6):
            su_2+=1
        if popu < 10*(10**6):
            su_3+=1
    return su_1,su_2,su_3

def find_c_and_5(data):
    su_c,su_5 = 0,0
    for country in data:
        if country.get('country')[0] == 'C':
            su_c += 1
        if len(country.get('country')) > 5:
            su_5+=1
    return su_c,su_5

def find_top20_top150(data):
    su = 0
    for i in range(20):
        su+=float(data[i].get('World'))
    per_1 = su
    su = 0
    for i in range(49,151):
        su+=float(data[i].get('World'))
    per_2 = su
    return per_1,per_2

def main():
    path = input('File Name: ')
    #path = 'python\data\worldpopulation.json'
    data = get_data(path)
    op = input('Input : ')
    if op == '1':
        print(len(data))
    elif op == '2':
        total = get_total_population(data)
        print(total)
    elif op == '3':
        su_c,su_5 = find_c_and_5(data)
        print(su_c)
        print(su_5)
    elif op == '4':
        su_1,su_2,su_3 = fliter_population(data)
        print(su_1)
        print(su_2)
        print(su_3)
    elif op == '5':
        per_1,per_2 = find_top20_top150(data)
        print(f'{per_1*100:.2f}')
        print(f'{per_2*100:.2f}')
    
main()
```

## 10.L14 twitter

```python
import json
def get_data(path):
    with open(path,'r') as f:
        data = json.loads(f.read())
    return data

def check_people(data,key_check):
    tweet_name = list()
    su =0
    for tweet in data:
        if not(tweet.get(key_check) in tweet_name):
            tweet_name.append(tweet.get(key_check))
            su+=1
    return su

def check_priority_topic(data,key_check):
    su=0
    for tweet in data:
        if tweet.get('topic_priority') == key_check:
            su+=1
    return su

def check_maxtweet(data):
    tweet_name = dict()
    for tweet in data:
        if not(tweet.get('author') in tweet_name):
            tweet_name[tweet.get('author')] = 1
        else:
            tweet_name[tweet.get('author')]+=1
    max_ = max(tweet_name.values())
    max_list = []
    for name,n in tweet_name.items():
        if n == max_:
            max_list.append(name)
    return max_list

def check_language(data):
    for tweet in data:
        if tweet.get('language') != 'EN':
            return True
    return False
    
def check_word(data):
    max_ = 0
    for tweet in data:
        word_count = len(tweet.get('text').split(' '))
        if word_count > max_:
            max_ = word_count
    return max_

def main():
    path = input('File name: ')
    #path = 'python\data\\twitter.json'
    data = get_data(path)
    op = input('input: ')
    if op == '1':
        print(len(data))
    elif op == '2':
        print(check_people(data,'author'))
    elif op == '3':
        name_list = check_maxtweet(data)
        for name in name_list:
            print(name)
    elif op == '4':
        print(check_people(data,'topic'))
    elif op == '5':
        print(check_priority_topic(data,'ALERT'))
    elif op == '6':
        print(check_priority_topic(data,'UNIMPORTANT'))
    elif op == '7':
        print(check_language(data))
    elif op == '8':
        print(check_word(data))
    
main()
```
